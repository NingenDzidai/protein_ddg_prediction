```python
import pandas as pd
import os
import pathlib
import numpy as np
```

```python
# search for dataset folder
path = pathlib.Path(os.getcwd())
path = path.parent.absolute()
path_data = os.path.join(path, 'data')
path = os.path.join(path, 'data', 'skempi_v2.csv')
print(path)
```

```python
# Load CSV-file SKEMPI 2.0.
df = pd.read_csv(path, delimiter=';')
```

```python
df
```

```python
# Non-useful columns dropped 
columns_to_drop = [
    '#Pdb', 'Mutation(s)_PDB', 'Hold_out_proteins', 
    'Affinity_mut (M)', 'Affinity_wt (M)', 'Reference',
    'kon_mut (M^(-1)s^(-1))', 'kon_mut_parsed', 'kon_wt (M^(-1)s^(-1))', 'kon_wt_parsed',
    'koff_mut (s^(-1))', 'koff_mut_parsed', 'koff_wt (s^(-1))', 'koff_wt_parsed',
    'dH_mut (kcal mol^(-1))', 'dH_wt (kcal mol^(-1))', 'dS_mut (cal mol^(-1) K^(-1))', 'dS_wt (cal mol^(-1) K^(-1))',
    'Notes', 'Method', 'SKEMPI version'
]
df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
# Deletes rows with missing values in key columns Affinity_mut_parsed, Affinity_wt_parsed
df['Affinity_mut_parsed'] = pd.to_numeric(df['Affinity_mut_parsed'], errors='coerce')
df['Affinity_wt_parsed'] = pd.to_numeric(df['Affinity_wt_parsed'], errors='coerce')
df.dropna(subset=['Affinity_mut_parsed', 'Affinity_wt_parsed'], inplace=True)

# Clean the Temperature column from non-numeric values
df['Temperature'] = df['Temperature'].str.replace(r'[^0-9.]', '', regex=True)  # Keep only numbers
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
df.dropna(subset=['Temperature'], inplace=True)
df = df.reset_index(drop=True)
```

```python
df
```

```python
# Counting the number of values in cells
def calculate_values(dataframe, column_name):
    counts = dataframe[column_name].str.split(",").apply(len)
    distribution = counts.value_counts().sort_index()
    print(distribution)

calculate_values(df, "Mutation(s)_cleaned")
```

```python
# In case there are many values inside the cell, we delete those that do not carry much value for training
def drop_rows(dataframe, column_name, to_drop):
    counts = dataframe[column_name].str.split(",").apply(len)
    dataframe_filtered = dataframe[counts <= to_drop].reset_index(drop=True)
    print(f"Dropped {len(dataframe) - len(dataframe_filtered)} rows. Remaining: {len(dataframe_filtered)}")
    return dataframe_filtered

df_filtered = drop_rows(df, "Mutation(s)_cleaned", 2)
df_filtered
```

```python
# See the data in the Hold_out_type column 
df['Hold_out_type'] = df['Hold_out_type'].where(df['Hold_out_type'].notna(), 'missing')
df['Hold_out_type'].unique()
```

```python
# Splitting multiple values in a cell
def column_split(dataframe, to_split):
    for col in to_split:
        split_df = dataframe[col].str.split(",", expand=True)
        split_df.columns = [f"{col}_{i+1}" for i in range(split_df.shape[1])]
        dataframe = pd.concat([dataframe, split_df], axis=1)
    return dataframe

to_split = ["Mutation(s)_cleaned", "iMutation_Location(s)"]
df_splitted = column_split(df_filtered, to_split)
df_splitted
```

```python
columns_to_drop = [
    'Mutation(s)_cleaned', 'iMutation_Location(s)'
]
df = df_splitted.drop(columns=columns_to_drop)
df
```

```python
def calculate_ddg(df):
    """
    Считаем ΔΔG и создаем таргет ddG_sign.
    """
    df = df.copy()
    R = 8.314  # J/(mol·K)

    # Translate into numerical format
    df["Affinity_mut_parsed"] = pd.to_numeric(df["Affinity_mut_parsed"], errors="coerce")
    df["Affinity_wt_parsed"] = pd.to_numeric(df["Affinity_wt_parsed"], errors="coerce")
    df["Temperature"] = pd.to_numeric(df["Temperature"], errors="coerce")

    mask = df["Affinity_mut_parsed"].notna() & df["Affinity_wt_parsed"].notna() & df["Temperature"].notna()

    # Calculate ΔΔG
    T_kelvin = df.loc[mask, "Temperature"]
    RT_kcal = (R * T_kelvin)

    df.loc[mask, "ddG_kcal"] = RT_kcal * np.log(df.loc[mask, "Affinity_mut_parsed"] / df.loc[mask, "Affinity_wt_parsed"])

    # Create binary target
    df["ddG_sign"] = (df["ddG_kcal"] > 0).astype(int)

    return df
```

```python
df = calculate_ddg(df)
df
```

```python
def parse_and_replace_mutation_columns(df, prefix="Mutation(s)_cleaned_", max_mutations=5):
    """
    Разделение колонн типа 'Mutation(s)_cleaned_<N>' на:
    orig_aa_N, chain_N, residue_num_N, mut_aa_N
    """
    df = df.copy()
    
    for i in range(1, max_mutations + 1):
        col = f"{prefix}{i}"
        if col not in df.columns:
            continue

        # Extract components using regex
        parsed = df[col].dropna().astype(str).str.extract(r'^([A-Z])([A-Z])(\d+[A-Z]?)([A-Z])$')
        parsed.columns = [f"orig_aa_{i}", f"chain_{i}", f"residue_num_{i}", f"mut_aa_{i}"]

        # Join parsed columns into DataFrame
        df = df.drop(columns=[col]).join(parsed)

    df = df.where(pd.notnull(df), "None")
    
    return df
```

```python
df = parse_and_replace_mutation_columns(df)
to_drop = ["Affinity_mut_parsed", "Affinity_wt_parsed", "Temperature", "ddG_kcal"]
df = df.drop(columns=to_drop)
```
